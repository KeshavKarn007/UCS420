```python
import numpy as np
import matplotlib.pyplot as plt
M = float(input("Enter a value for M: "))
x = np.linspace(-10, 10, 400)
y1 = M * x ** 2
y2 = M * np.sin(x)
plt.figure(figsize=(10, 6))
plt.plot(x, y1, label=r'$y = M \cdot x^2$', color='b', linestyle='-')
plt.plot(x, y2, label=r'$y = M \cdot \sin(x)$', color='r', linestyle='--')
plt.title(f"Plot of y = {M} * x² and y = {M} * sin(x)")
plt.xlabel("x-axis")
plt.ylabel("y-axis")
plt.legend()
plt.grid()
plt.show()
print(f"You entered M = {M}")
```

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

df = pd.DataFrame({'Subject': ['Math', 'Science', 'English', 'History', 'Art'], 'Score': [85, 90, 78, 88, 92]})

plt.figure(figsize=(8, 5))
ax = sns.barplot(x='Subject', y='Score', data=df, hue='Subject', dodge=False, palette='husl', legend=False)

for i, row in df.iterrows():
    ax.text(i, row['Score'] + 1, str(row['Score']), ha='center', fontsize=12)

plt.title('Scores in Different Subjects')
plt.xlabel('Subjects')
plt.ylabel('Scores')
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

roll_number = int(input("Enter your roll number: "))
np.random.seed(roll_number)
data = np.random.randn(50)

fig, axes = plt.subplots(2, 2, figsize=(10, 8))

axes[0, 0].plot(np.cumsum(data), color='b', linestyle='-')
axes[0, 0].set_title("Cumulative Sum")
axes[0, 0].set_xlabel("Index")
axes[0, 0].set_ylabel("Cumulative Sum")
axes[0, 0].grid(True)

axes[0, 1].scatter(range(50), data, color='r', alpha=0.7)
axes[0, 1].set_title("Scatter Plot with Noise")
axes[0, 1].set_xlabel("Index")
axes[0, 1].set_ylabel("Value")
axes[0, 1].grid(True)

fig.delaxes(axes[1, 0])
fig.delaxes(axes[1, 1])

plt.tight_layout()
plt.show()
```

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

df = pd.read_csv("company_sales_data.csv")

plt.figure(figsize=(10, 5))
sns.lineplot(x=df.columns[0], y='total_profit', data=df, marker='o', color='b')
plt.title('Total Profit Over Months')
plt.xlabel(df.columns[0])
plt.ylabel('Total Profit')
plt.grid(True)
plt.show()

plt.figure(figsize=(12, 6))
for col in df.columns[1:-1]:
    sns.lineplot(x=df.columns[0], y=col, data=df, label=col)
plt.title('Sales Data for All Products')
plt.xlabel(df.columns[0])
plt.ylabel('Sales')
plt.legend()
plt.grid(True)
plt.show()

df.set_index(df.columns[0]).plot(kind='bar', figsize=(12, 6), colormap='tab10')
plt.title('Sales Data for All Features')
plt.xlabel(df.columns[0])
plt.ylabel('Values')
plt.grid(True, axis='y')
plt.show()
```
