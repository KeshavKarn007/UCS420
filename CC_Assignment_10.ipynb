```python
import re
import nltk
from nltk.corpus import stopwords
from nltk.tokenize import sent_tokenize, word_tokenize
from collections import Counter

# Download required NLTK resources
nltk.download('punkt')
nltk.download('punkt_tab')  # For newer NLTK versions
nltk.download('stopwords')

# 1. Unique paragraph (about books)
paragraph = (
    "Books have always been my window to the world. "
    "They allow me to travel to distant lands, explore new ideas, and meet fascinating characters. "
    "A good book can inspire, educate, and entertain all at once. "
    "I especially enjoy reading mysteries and historical fiction. "
    "For me, books are not just stories; they are lifelong friends."
)

# 2. Convert to lowercase and remove punctuation using re
paragraph_clean = re.sub(r'[^\w\s]', '', paragraph.lower())

# 3. Tokenize into sentences and words
sentences = sent_tokenize(paragraph_clean)
words_nltk = word_tokenize(paragraph_clean)

# 4. Split using split() and compare with word_tokenize()
words_split = paragraph_clean.split()
print("Split using split():")
print(words_split)
print("\nSplit using word_tokenize():")
print(words_nltk)
print("\nDifference (tokens in word_tokenize() but not in split()):")
print(set(words_nltk) - set(words_split))

# 5. Remove stopwords
stop_words = set(stopwords.words('english'))
filtered_words = [word for word in words_nltk if word not in stop_words]

# 6. Display word frequency distribution (excluding stopwords)
freq_dist = Counter(filtered_words)
print("\nWord Frequency Distribution (excluding stopwords):")
for word, freq in freq_dist.most_common():
    print(f"{word}: {freq}")

# Optional: Print tokenized sentences
print("\nTokenized Sentences:")
for i, sent in enumerate(sentences, 1):
    print(f"{i}: {sent}")
```

```python
import re
import nltk
from nltk.tokenize import word_tokenize
from nltk.corpus import stopwords, wordnet
from nltk.stem import PorterStemmer, WordNetLemmatizer

# Download required resources
nltk.download('punkt')
nltk.download('stopwords')
nltk.download('wordnet')
nltk.download('omw-1.4')

# Paragraph from Q1
text = """I love exploring new advancements in technology. Artificial Intelligence, Machine Learning, and Robotics fascinate me the most.
Every year, new gadgets and software redefine what's possible in the tech world. I enjoy reading tech blogs and watching documentaries
about the future of innovation. It's amazing how quickly technology evolves and impacts every aspect of our lives."""

# Step 1: Extract all words with only alphabets using re.findall()
words_alpha = re.findall(r'\b[a-zA-Z]+\b', text.lower())

# Step 2: Remove stop words
stop_words = set(stopwords.words('english'))
filtered_words = [word for word in words_alpha if word not in stop_words]

# Step 3: Perform stemming with PorterStemmer
stemmer = PorterStemmer()
stemmed_words = [stemmer.stem(word) for word in filtered_words]

# Step 4: Perform lemmatization with WordNetLemmatizer
lemmatizer = WordNetLemmatizer()
lemmatized_words = [lemmatizer.lemmatize(word) for word in filtered_words]

# Step 5: Compare stemmed and lemmatized outputs
print("\nFiltered words (without stopwords):")
print(filtered_words)

print("\nStemmed words:")
print(stemmed_words)

print("\nLemmatized words:")
print(lemmatized_words)

# Step 5 Explanation
print("\nComparison:")
print("""\nStemming is faster but may produce non-real words (e.g., 'technolog' from 'technology').
Lemmatization is slower but gives meaningful base forms (e.g., 'running' → 'run').
Use stemming when speed matters and exact form isn't important.
Use lemmatization when meaning and grammatical correctness are required, such as in NLP tasks.""")

```

```python
from sklearn.feature_extraction.text import CountVectorizer, TfidfVectorizer
import numpy as np

# 3 short texts (news headlines/product reviews)
texts = [
    "Apple unveils new iPhone with advanced camera features.",
    "Scientists discover water on Mars in groundbreaking study.",
    "The new electric car offers impressive range and fast charging."
]

# 1. Bag of Words with CountVectorizer
count_vect = CountVectorizer()
bow = count_vect.fit_transform(texts)
print("Bag of Words Representation (CountVectorizer):")
print(count_vect.get_feature_names_out())
print(bow.toarray())

# 2. TF-IDF with TfidfVectorizer
tfidf_vect = TfidfVectorizer()
tfidf = tfidf_vect.fit_transform(texts)
feature_names = tfidf_vect.get_feature_names_out()

print("\nTF-IDF Scores:")
print(tfidf.toarray())

# 3. Top 3 keywords per text using TF-IDF
print("\nTop 3 Keywords per Text (by TF-IDF):")
for i, row in enumerate(tfidf.toarray()):
    top_indices = row.argsort()[-3:][::-1]
    keywords = [feature_names[idx] for idx in top_indices]
    print(f"Text {i+1}: {keywords}")

# Interpretation Example
print("\nInterpretation:")
print("TF-IDF highlights the most unique and important words in each text. For example, 'iphone' and 'camera' are top keywords for the first text, indicating the focus on Apple's new product features.")
```

```python
import re
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.metrics.pairwise import cosine_similarity

# 2 short texts describing different technologies
text1 = """Artificial Intelligence (AI) refers to the simulation of human intelligence in machines.
AI systems can analyze data, recognize patterns, and make decisions with minimal human intervention.
It is widely used in applications like speech recognition, image analysis, and autonomous vehicles."""

text2 = """Blockchain is a decentralized digital ledger that records transactions across multiple computers.
It ensures transparency and security by making data tamper-resistant.
Blockchain technology is the backbone of cryptocurrencies and is also used in supply chain management."""

# Preprocess: lowercase, remove punctuation, tokenize
def preprocess(text):
    text = text.lower()
    text = re.sub(r'[^\w\s]', '', text)
    tokens = text.split()
    return tokens

tokens1 = preprocess(text1)
tokens2 = preprocess(text2)

# a. Jaccard Similarity
set1 = set(tokens1)
set2 = set(tokens2)
intersection = set1.intersection(set2)
union = set1.union(set2)
jaccard_sim = len(intersection) / len(union)
print(f"Jaccard Similarity: {jaccard_sim:.4f}")

# b. Cosine Similarity using TfidfVectorizer
tfidf_vect = TfidfVectorizer()
tfidf_matrix = tfidf_vect.fit_transform([text1, text2])
cos_sim = cosine_similarity(tfidf_matrix[0:1], tfidf_matrix[1:2])[0][0]
print(f"Cosine Similarity: {cos_sim:.4f}")

# c. Analysis
print("\nAnalysis:")
print("Jaccard Similarity is low because the two texts use mostly different words, reflecting their different topics.")
print("Cosine Similarity is also low, but it considers term frequency and the importance of words, offering a more nuanced view.")
print("In this case, Cosine Similarity provides better insight into the overall thematic difference between AI and Blockchain, as it captures the context and importance of words, not just their overlap.")
```

```python
from textblob import TextBlob
from wordcloud import WordCloud
import matplotlib.pyplot as plt

# Write a short review for a product or service
reviews = [
    "This smartphone has an amazing camera and battery life.",
    "The delivery was late and the packaging was damaged.",
    "Customer service was helpful and resolved my issue quickly.",
    "The product did not work as expected and I am disappointed.",
    "Absolutely love this laptop! Fast, lightweight, and reliable."
]

# 1. Use TextBlob for polarity & subjectivity
polarities = []
subjectivities = []
sentiments = []

for review in reviews:
    blob = TextBlob(review)
    pol = blob.sentiment.polarity
    sub = blob.sentiment.subjectivity
    polarities.append(pol)
    subjectivities.append(sub)
    if pol > 0.1:
        sentiments.append("Positive")
    elif pol < -0.1:
        sentiments.append("Negative")
    else:
        sentiments.append("Neutral")
    print(f"Review: {review}\n  Polarity: {pol:.2f}, Subjectivity: {sub:.2f}, Sentiment: {sentiments[-1]}\n")

# 2. Create word cloud for all positive reviews
positive_reviews = " ".join([rev for rev, sent in zip(reviews, sentiments) if sent == "Positive"])
if positive_reviews:
    wordcloud = WordCloud(width=600, height=400, background_color='white').generate(positive_reviews)
    plt.figure(figsize=(8, 5))
    plt.imshow(wordcloud, interpolation='bilinear')
    plt.axis('off')
    plt.title("Word Cloud for Positive Reviews")
    plt.show()
else:
    print("No positive reviews to generate word cloud.")

```

```python
import numpy as np
import tensorflow as tf
from tensorflow.keras.preprocessing.text import Tokenizer
from tensorflow.keras.utils import to_categorical
from tensorflow.keras.preprocessing.sequence import pad_sequences
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Embedding, LSTM, Dense

# 1. Choose a paragraph (~100 words) as training data
data = """
Artificial intelligence has transformed our modern world in countless ways. From smartphones to smart homes, AI algorithms power the devices we use every day. These systems analyze vast amounts of data to recognize patterns and make predictions with impressive accuracy. Machine learning models continue to improve as they process more information, becoming increasingly sophisticated in their capabilities. Natural language processing allows computers to understand and generate human language, enabling virtual assistants to respond to our questions. Computer vision systems can identify objects, faces, and even emotions in images and video. As AI technology advances, it raises important questions about privacy, ethics, and the future relationship between humans and machines.
"""

# 2. Tokenize text using Tokenizer from keras.preprocessing.text
tokenizer = Tokenizer()
tokenizer.fit_on_texts([data])
vocab_size = len(tokenizer.word_index) + 1

# 3. Create input sequences
encoded = tokenizer.texts_to_sequences([data])[0]
sequences = []
for i in range(2, len(encoded)):
    seq = encoded[i-2:i+1]
    sequences.append(seq)
sequences = np.array(sequences)

X, y = sequences[:,:-1], sequences[:,-1]
y = to_categorical(y, num_classes=vocab_size)

# 4. Build a simple LSTM model
model = Sequential()
model.add(Embedding(input_dim=vocab_size, output_dim=10, input_length=2))
model.add(LSTM(50))
model.add(Dense(vocab_size, activation='softmax'))
model.compile(loss='categorical_crossentropy', optimizer='adam', metrics=['accuracy'])
print("Model compiled successfully!")

# 5. Train the model
model.fit(X, y, epochs=100, verbose=1)

# 6. Generate new text from seed words
def generate_text(seed_text, n_words):
    result = seed_text.split()
    for _ in range(n_words):
        # Convert the text to a sequence
        token_list = tokenizer.texts_to_sequences([seed_text])[0]
        # Take the last 2 words as input
        token_list = token_list[-2:]
        # Pad the sequence
        token_list = pad_sequences([token_list], maxlen=2, padding='pre')
        # Predict the next word
        predicted = np.argmax(model.predict(token_list, verbose=0), axis=-1)

        # Convert predicted token back to word
        output_word = ""
        for word, index in tokenizer.word_index.items():
            if index == predicted:
                output_word = word
                break

        # Add to result and update seed text
        result.append(output_word)
        seed_text = " ".join(result[-2:])

    return " ".join(result)

# Generate text from different seed words
print("\nGenerated text from 'artificial intelligence':")
print(generate_text("artificial intelligence", 10))

print("\nGenerated text from 'machine learning':")
print(generate_text("machine learning", 10))
```
